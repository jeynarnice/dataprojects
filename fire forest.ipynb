```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11n.pt")

# Train the model
train_results = model.train(
    data= "/content/drive/MyDrive/Colab Notebooks/ml/forest fire.v5i.yolov11/data.yaml",  # path to dataset YAML
    epochs=100,  # number of training epochs
    imgsz=640,  # training image size
    device=0,  # use GPU; '0' specifies the first GPU   or device=0,1,2,3 or device=cpu
    workers=2  # use multiple workers for data loading
)

# Evaluate model performance on the validation set
metrics = model.val()
```
